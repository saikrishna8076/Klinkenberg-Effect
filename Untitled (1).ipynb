```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Constants for the core sample and gas properties
mu = 0.018  # gas viscosity in cP
L = 5.0     # length of core sample in cm
A = 10.0    # cross-sectional area in cmÂ²

# Experimental Observations (Example Values)
data = {
    'DeltaP_atm': [0.25, 0.50, 1.00],     # Pressure difference P1 - P0
    'Qi_cc_sec': [150, 200, 300],         # Observed flow rate in cc/sec
    'P0_atm': [1.0, 1.0, 1.0]             # Outlet pressure
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate P1 (upstream pressure)
df['P1'] = df['DeltaP_atm'] + df['P0_atm']

# Calculate Mean Pressure (Pm)
df['Pm'] = (df['P1'] + df['P0_atm']) / 2

# Corrected flow rate Qm = (P1 * Qi) / Pm
df['Qm'] = (df['P1'] * df['Qi_cc_sec']) / df['Pm']

# Calculate Ka using Darcy's Law: Ka = (Qm * mu * L) / (A * DeltaP)
df['Ka_darcy'] = (df['Qm'] * mu * L) / (A * df['DeltaP_atm'])

# Calculate 1/Pm for Klinkenberg plot
df['1/Pm'] = 1 / df['Pm']

# Linear Regression for Klinkenberg Plot (Ka vs 1/Pm)
X = df['1/Pm'].values
Y = df['Ka_darcy'].values
slope, intercept = np.polyfit(X, Y, 1)

KL = intercept  # True Permeability (absolute)
b_factor = slope  # Klinkenberg factor

# Print Results
print("=== RESULTS ===")
print(df)
print(f"\nTrue Permeability (KL): {KL:.4f} Darcy")
print(f"Klinkenberg Factor (b): {b_factor:.4f}")

# Visualization
sns.set(style="whitegrid")
plt.figure(figsize=(8,6))
sns.scatterplot(x='1/Pm', y='Ka_darcy', data=df, color='blue', s=80, label='Observed Data')
plt.plot(df['1/Pm'], slope * df['1/Pm'] + intercept, color='red', label='Regression Line')

plt.title('Klinkenberg Plot: Ka vs 1/Pm')
plt.xlabel('1 / Mean Pressure (1/atm)')
plt.ylabel('Gas Permeability Ka (Darcy)')
plt.legend()
plt.show()
```
